```python
balance=1000
def deposit(amount):
    global balance
    balance += amount
    print(f"Deposited ₹{amount}. New Balance: ₹{balance}")

def withdraw(amount):
    global balance
    if amount > balance:
        print(" Withdrawal failed! Insufficient balance.")
    else:
        balance -= amount
        print(f"Withdrew ₹{amount}. New Balance: ₹{balance}")

def check_balance():
    print(f"Current Balance: ₹{balance}")

deposit(4000)
withdraw(2000)
withdraw(1000)
check_balance()
```

```python
temperatures = [33, 27, 33, 28, 39, 31]
humidity_values = [60, 85, 70, 75, 65, 68]

def average_temperature(data):
    return sum(data) / len(data)

def average_humidity(data):
    return sum(data) / len(data)

def weather_condition(temperature, humidity):
    if temperature > 30:
        print("Weather Condition: Sunny")
    elif temperature < 25:
        print("Weather Condition: Rainy")
    else:
        print("Weather Condition: Pleasant")

    if humidity > 70:
        print("Humidity Status: Very Humid")
    elif humidity < 20:
        print("Humidity Status: Dry")
    else:
        print("Humidity Status: Normal Air")

avg_temperature = average_temperature(temperatures)
avg_humidity = average_humidity(humidity_values)

print(f"Average Temperature: {avg_temperature:.2f} °C")
print(f"Average Humidity: {avg_humidity:.2f}%")
weather_condition(avg_temperature, avg_humidity)
```
