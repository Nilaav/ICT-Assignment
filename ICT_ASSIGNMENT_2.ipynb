```python
import numpy as np
import pandas as pd

df = pd.read_csv('auto-mpg (1).csv')
df.head()
```

```python
df['horsepower'] = df['horsepower'].replace('?', np.nan).astype(float)
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].mean())
```

##NumPy Problems

```python
#1
mpg = df['mpg'].to_numpy()

mean_mpg = mpg.mean()
median_mpg = np.median(mpg)
std_mpg = mpg.std()
count_gt_25 = np.sum(mpg > 25)

mean_mpg, median_mpg, std_mpg, count_gt_25
```

```python
#2
cars_gt_6 = df[df['cylinders'] > 6]['car name'].to_numpy()
cars_gt_6
```

```python
#3
weight = df['weight'].to_numpy()
np.percentile(weight, [25, 50, 75])
```

```python
#4
acc = df['acceleration'].to_numpy()
acc_normalized = (acc - acc.min()) / (acc.max() - acc.min())
acc_normalized[:5]
```

```python
#5
horsepower_updated = df['horsepower'].to_numpy() * 1.1
horsepower_updated[:5]
```

```python
#6
avg_disp_europe = df[df['origin'] == 2]['displacement'].to_numpy().mean()
avg_disp_europe
```

```python
#7
matrix = df[['mpg', 'horsepower', 'weight']].to_numpy()
vector = np.array([1, 0.5, -0.2])

dot_product = matrix.dot(vector)
dot_product[:5]
```

```python
#8
sorted_cars = df.sort_values('model year', ascending=False)
sorted_cars['car name'].head(5)
```

```python
#9
correlation = np.corrcoef(df['mpg'], df['weight'])[0,1]
correlation
```

```python
#10
cylinders = df['cylinders'].to_numpy()
mpg = df['mpg'].to_numpy()

for c in np.unique(cylinders):
    print(c, mpg[cylinders == c].mean())
```

##Pandas Problems

```python
#1
df.head(10)
df.shape
df.describe()
```

```python
#2
df_1975 = df[(df['model year'] == 75) & (df['weight'] < 3000)]
df_1975[['car name', 'weight', 'mpg']]
```

```python
#3
df.isnull().sum()
```

```python
#4
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
df.head()
```

```python
#5
df.groupby('origin')['mpg'].mean()
```

```python
#6
df.sort_values('mpg', ascending=False)[['car name', 'mpg']].head(10)
```

```python
#7
def performance_score(row):
    return (row['mpg'] * row['acceleration']) / row['weight']

df['performance_score'] = df.apply(performance_score, axis=1)
df.head()
```

```python
#8
year_summary = df.groupby('model year')[['mpg','weight','horsepower']].mean()
year_summary
```

```python
#9
high_mpg = df[df['mpg'] > 30][['mpg','cylinders','horsepower','weight']]
high_mpg.to_csv('high_mpg_cars.csv', index=False)
```
